```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib
```

```python
students=pd.read_csv("Placement_Data_Full_Class.csv")
```

```python
students.info()
```

```python
students.describe()
```

```python
students.head()
```

```python
students.drop(['salary','Sno','Mba %','specialisation'],axis=1,inplace=True)
```

```python
le=LabelEncoder()
```

```python
students['Degree stream']=le.fit_transform(students['Degree stream'])
```

```python
students['12th Stream']=le.fit_transform(students['12th Stream'])
```

```python
students['Gender']=le.fit_transform(students['Gender'])
```

```python
students['SSC Board']=le.fit_transform(students['SSC Board'])
```

```python
students['HSC Board']=le.fit_transform(students['HSC Board'])
```

```python
students['status']=le.fit_transform(students['status'])
```

```python
students['Work exp']=le.fit_transform(students['Work exp'])
```

```python
students.corr(numeric_only=True)['status'].sort_values()
```

```python
X=students.drop('status',axis=1)
y=students['status']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train
```

```python
model=LogisticRegression(max_iter=200)  # max_iter sets the maximum number of iterations to converge
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
joblib.dump(model, 'logistic_regression_model.joblib')
```
